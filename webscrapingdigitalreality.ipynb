```python
import requests
from lxml import html

# URL of the webpage
url = 'https://www.digitalrealty.com/data-centers/emea/london'

# Fetch the webpage
response = requests.get(url)

print(response)

# Check if request was successful
if response.status_code == 200:
    # Parse the HTML content
    tree = html.fromstring(response.text)

    # Use XPath to select elements
    elements = tree.xpath('//*[@id="__NEXT_DATA__"]')

    # Do something with the selected elements
    for element in elements:
        print(html.tostring(element))
else:
    print(f"Failed to fetch webpage. Status code: {response.status_code}")
```

```python
import requests
import json
from lxml import html

# URL of the webpage
url = 'https://www.digitalrealty.com/data-centers/emea/london'

# Fetch the webpage
response = requests.get(url)

# Check if request was successful
if response.status_code == 200:
    # Parse the HTML content
    tree = html.fromstring(response.text)

    # Use XPath to select the element with id="__NEXT_DATA__"
    element = tree.xpath('//*[@id="__NEXT_DATA__"]/text()')

    if element:
        # Load the JSON content from the element
        json_data = json.loads(element[0])

        # Define a recursive function to print JSON data hierarchically
        def print_json_hierarchy(data, indent=0):
            indent_str = '  ' * indent
            if isinstance(data, dict):
                for key, value in data.items():
                    print(f"{indent_str}{key}:")
                    print_json_hierarchy(value, indent + 1)
            elif isinstance(data, list):
                for index, item in enumerate(data):
                    print(f"{indent_str}[{index}]:")
                    print_json_hierarchy(item, indent + 1)
            else:
                print(f"{indent_str}{data}")

        # Print the JSON data hierarchically
        print_json_hierarchy(json_data)
    else:
        print("Element with id='__NEXT_DATA__' not found.")
else:
    print(f"Failed to fetch webpage. Status code: {response.status_code}")

```
